# Religon and Income Mapped

In this notebook, we correlate county-level religious data with county-level per-capita income data to see if there is any statistical relationship between these two variables. Then, we visualize the data on a map of the United States and show where certain religions are the most common to get a better understanding of religion in the country. This is important because it could help politicians with their political agendas, determine the characteristics of a place, neighborhood, or country, and to just get a better understanding of people's political ideologies in the United States. This could also help influence public policy as the government could support religious populations that aren't succeeding quite as much as others in today’s society.

Computational methods are necessary because you need to analyze large data sets to obtain correct results. For example, both datsets we use to use are over 3,000 rowns in Excel. Python will be able to import this data and be able to easily obtain values that might not be possible in Excel. For the data visualization, we use geopandas to get a map of the United States that shows the population of any inputted religion in every county. This allows us to more accurately and easily analyze the many different religions that make up the United States

We use data from the [Longitudinal Religious Congregations and Membership File, 2010](http://www.thearda.com/Archive/Files/Downloads/RCMSCY10_DL2.asp) and [Unemployment and median household income for the U.S., States, and counties, 2007-18](https://www.ers.usda.gov/data-products/county-level-data-sets/download-data/) to help answer the question. This material is appropriate because the datasets give religion and household income data on every single county in the country. One issue with the religion data is that there are large subsections of the excel data that contain empty cells and there are a large amount of religions that have very few followers and don’t fall into the main religions in the world. What we decided to do was find the largest 10 religions in the United States, and then make a new dataframe with just those 10 religions. This makes the data slightly easier to work with, and narrows our focus so we aren't analyzing every religion in the world. The religion data didn't really need to be cleaned, it is in an easy to read excel csv file that imported nicely to python. The income data, however, did need some cleaning. We had to remove dollar signs and commas so python recognized that they were numbers and didn't give an error message back everytime. Luckily, we are able to connect the two databases for easy analysis because they both have a FIPS code which is basically a county identifier for each county in the country. This ensured that the data was merging correctly.


Here is a little cheatsheet to help understand the datasets better

Dataset | Use | Shape | Examples
-----| :----- | :---- | ---
Religion_data | data on how many adherents for every single different type of religion that exists in the country (also subsects of religions) | 3,000 rows in total. | Lists county population, county name, state, and adherents to religions.
Income_data | Economic statistics for each county in the US | 3,000 rows in total. | Focuses on labor employment and household income in the county. We decide to focus on the Median_Household_Income_2017 column for our analysis

Importing the proper libraries

```python
# Importing the proper libraries
import pandas as pd
import numpy as np

import geopandas as gp
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression 

import statsmodels.formula.api as smf
from statsmodels.api import add_constant
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

There are 3 datasets we will be looking at. The income dataset, the relgion dataset, and the GeoJson for the counties. (We alsoe have a fips table for county work, although it is not used in the current version.)

The first 3 datasets are coming off of Github, however the shape files have to be downloaded to work. Here's a link to download the county shape dataset. [TIGER/line from Data.Gov](https://catalog.data.gov/dataset/tiger-line-shapefile-2017-nation-u-s-current-county-and-equivalent-national-shapefile)

```python
religion_data = pd.read_stata("https://raw.githubusercontent.com/Book81able/DS2000FinalProject/master/religon.DTA")
income_data = pd.read_csv("https://raw.githubusercontent.com/Book81able/DS2000FinalProject/master/unemployment.csv", skipinitialspace=True)
fips = pd.read_csv("https://raw.githubusercontent.com/kjhealy/fips-codes/master/state_and_county_fips_master.csv")

geo = gp.read_file("c_02ap19/c_02ap19.shp")
```

```python
religion_data.head()
```

```python
income_data.head()
```

Now we need to remove the dollar signs from income_data median income row so we can analyze it.

```python
cols = ['Median_Household_Income_2017']
for item in cols:
  income_data[item] = income_data[item].str.replace('$','')
  income_data[item] = income_data[item].str.replace(',','')
  income_data[item] = pd.to_numeric(income_data[item])
income_data.rename(columns={"FIPS":'fips'},inplace=True)
income_data.head()
```

To make getting the largest religions easier we're temporarily dropping these columnns that have large values.


```python
religion_data_dropped = religion_data.drop(columns = ['POP2010', 'totadh', 'fips', 'totrate'])
religion_data_dropped.head()
```

Find the 10 Largest Religions

```python
x = religion_data_dropped.sum(numeric_only=True)
y = x.nlargest(11)
y = y[1:]
y
```

Make a new DataFrame with just these specific columns

```python
topreligionsdf = y.to_frame().reset_index()
topreligionsdf
```

Transpose this DataFrame so it looks more like the original dataframe.

We're also adding a few other religous groups from non-Christan traditions for more analysis.

```python
topreligionstransposed = topreligionsdf.transpose()
newframe = topreligionstransposed.rename(columns=topreligionstransposed.iloc[0])
#remove first row
newframe_dropped = newframe.drop('index')
newframe_dropped
a = newframe_dropped.columns.values
type(a)
a_list = a.tolist()
a_list.extend(['mslmadh','ojudadh','rfrmadh','budmadh','hniadh','zoroadh'])
a_list
```

For labeling we can also get the names of columns in a readable format. These we're hand copied from the lookup file that came with the religion data.

```python
religion_list = ['Catholic', 'Evangelical Protestant', 'Mainline Protestant', 'Southern Baptist Convention', 'Other', 'Non-denominational', 'United Methodist', 'Church of Jesus Christ of Latter-day Saints', 'Black Protestant', 'Evangelical Lutheran',"Muslim",'Orthodox Judaism','Reform Judaism','Buddhism',"Hindu","Zoroatrianism"]
religion_list
```

We're now filtering out all the religions so we only have those found in the religion_list. 

```python
# Start the filtered list
religion_data_filtered = religion_data[a_list]
pd.options.mode.chained_assignment = None #get rid of annoying error message

# Bring in some other important bits of data
religion_data_filtered['stabbr'] = pd.Series(religion_data['stabbr'])
religion_data_filtered['cntyname'] = pd.Series(religion_data['cntyname'])
religion_data_filtered['POP2010'] = pd.Series(religion_data['POP2010'])
religion_data_filtered['totadh'] = pd.Series(religion_data['totadh'])
religion_data_filtered['fips'] = pd.Series(religion_data["fips"])

# Add a column to track the percent one religon is in each county. (Based on total adherence)
for item in a_list:
  religion_data_filtered[item + "%"] = religion_data_filtered[item] / religion_data_filtered["totadh"]
religion_data_filtered.head()
```

Now we need to join income data and religion data by county code (fips)

```python
df_combined = pd.merge(religion_data_filtered, income_data, on='fips')

# For map reasons we're only going to be looking at the lower-48
df_combined = df_combined[df_combined.stabbr != 'AK']
df_combined = df_combined[df_combined.stabbr != 'HI']
df_combined.head()
```

# Data Analysis

Find all correlation values between religion population and Median Household Income

```python
correlations_list = []
for element in a_list:
  correlations = df_combined[str(element)].corr(df_combined['Median_Household_Income_2017'])
  correlations_list.append(correlations)
corr_dataframe = pd.DataFrame(correlations_list, religion_list)
corr_dataframe
```

```python
corr_rel = zip(correlations_list,religion_list)

corr_rel_s = sorted([(x,y) for x,y in corr_rel])
y = [x for x,y in corr_rel_s]
x = [y for x,y in corr_rel_s]
fig,ax = plt.subplots(figsize=(30,10))

ax.bar(x=x,height=y)

for tick in ax.xaxis.get_major_ticks()[1::2]:
    tick.set_pad(15)
```

This shows the correlation values, the higher the number it is means the relationship is stronger than lower values. All these correlation values are relatively low, but they still help explain income levels to some extent. We would need a regression to figure out to what extent that is exactly.

We also have a stat for percent of religious population in each county. This might give more detail to us.

```python
correlations_list_p = []
for element in a_list:
  correlations = df_combined[str(element + '%')].corr(df_combined['Median_Household_Income_2017'])
  correlations_list_p.append(correlations)
corr_dataframe = pd.DataFrame(correlations_list_p, religion_list)
corr_dataframe
```

```python
corr_rel = zip(correlations_list_p,religion_list)

corr_rel_s = sorted([(x,y) for x,y in corr_rel])
y = [x for x,y in corr_rel_s]
x = [y for x,y in corr_rel_s]
fig,ax = plt.subplots(figsize=(30,10))

ax.bar(x=x,height=y)

for tick in ax.xaxis.get_major_ticks()[1::2]:
    tick.set_pad(15)
```

These are better measures. We can see that for some religons there's a slight correlation downward (higher ratio leads to a lower median income.) In the map section we can see how these geographically comapre.

Run Regressions to figure out exactly how much one's specific religion determines their income levels. I decided to run regressions for the religion with the highest correlation (Mainline Protestant) and that with the lowest correlation (Southern Baptist Convention), though this could be done with any religion if you change the amount we are looking at in the Find the 10 Largest Religions section

```python
result = smf.ols(formula="Median_Household_Income_2017 ~ mprtadh", data=df_combined).fit(cov_type='HC1')
print(result.summary())
result = smf.ols(formula="Median_Household_Income_2017 ~ sbcadh", data=df_combined).fit(cov_type='HC1')
print(result.summary())
```

These R^2 values are pretty low for both regressions. For the first regression, one's religion explains 11.6% of their income levels, while for the second regression, it only explains 0.5%.

For one final thing, to more visually see the regressions, let's plot percentage of religion in each county vs median income for each religon.

```python
x = df_combined['Median_Household_Income_2017']

fig,axes = plt.subplots(figsize=(8,80),nrows=len(a_list))

# Goes through each religion to make the plot
for i in range(len(axes)):
    y = df_combined[str(a_list[i]+"%")]
    ax = axes[i]
    ax.scatter(x=x,y=y)
    ax.set_title(religion_list[i])
    ax.set_xlabel("Median Income")
    ax.set_ylabel("Percent of Population for " + religion_list[i])
```

We can see why the regression so low. These graphs don't look anything like a linear model, more like bell curves. As populations grow lower, the graphs grow less accurate. 

```python
# There is some degree of bellcurve in the income data.
plt.hist(x=df_combined['Median_Household_Income_2017'])
plt.set_ylabel("Count")
plt.set_xlabel("Median Household Income 2017")
```

# Visualzation

Let's combine the data with the geo data and clean it all up for the maps.

```python
df_combined = df_combined.fillna(0)
df_combined = df_combined.round(5)
geo = geo.rename({"FIPS": "fips"},axis=1)
geo = geo.astype({'fips':'int32'})
gdf = geo.merge(df_combined,on="fips")
gdf.head()
```

For the religions we're looking at, the most interesting graphic will be the heat maps of percentage of total county adherents

```python
fig, axes = plt.subplots(figsize = (12, len(a_list)*4),nrows=len(a_list))
for i in range(len(axes)):
    gdf.plot(column=str(a_list[i] + "%"),ax=axes[i],legend=True,cmap="magma")
    axes[i].set_title(religion_list[i])
```

There's so many interesting tid-bits here, but overall it just goes to show how regional religion is in the United States. 

Maps might not be the way to go with smaller religions, there's just not enough data to present any narrative of reigon.

Let's of course also plot income on our map. 

```python
fig,ax = plt.subplots(figsize=(12,4))

gdf.plot(ax=ax,column='Median_Household_Income_2017',legend=True,cmap="magma")
```

Centralized around cities, alhtough not as much as I would think. Still, the Wash-Bos area is really rich along with the Bay Area.

I was interested to see if the south had a noticable decline in income. It is one of the darker regions, however their are still bubbles of capital in the major cities. 

Finally let's look at population in 2010. We're going to augment the data here with a log transformation 

```python
fig,ax = plt.subplots(figsize=(12,4))
gdf.plot(ax=ax,column=np.log(gdf["POP2010"]),legend=True,cmap="magma")
```

If anything this map is really beautiful. On the east side the urban rural divide is really noticeable. However western counties are much less well defined.
